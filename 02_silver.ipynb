```python
# Notebook 02 - Silver
# Description: Clean and validate the raw data from Bronze layer

BRONZE_PATH = "/Volumes/olist_project/bronze/raw_data"

# Load all datasets

df_customers = spark.read.csv(f"{BRONZE_PATH}/olist_customers_dataset.csv", header=True, inferSchema=True)
df_orders = spark.read.csv(f"{BRONZE_PATH}/olist_orders_dataset.csv", header=True, inferSchema=True)
df_order_items = spark.read.csv(f"{BRONZE_PATH}/olist_order_items_dataset.csv", header=True, inferSchema=True)
df_payments = spark.read.csv(f"{BRONZE_PATH}/olist_order_payments_dataset.csv", header=True, inferSchema=True)
df_reviews = spark.read.csv(f"{BRONZE_PATH}/olist_order_reviews_dataset.csv", header=True, inferSchema=True)
df_products = spark.read.csv(f"{BRONZE_PATH}/olist_products_dataset.csv", header=True, inferSchema=True)
df_sellers = spark.read.csv(f"{BRONZE_PATH}/olist_sellers_dataset.csv", header=True, inferSchema=True)
df_geolocation = spark.read.csv(f"{BRONZE_PATH}/olist_geolocation_dataset.csv", header=True, inferSchema=True)
df_category = spark.read.csv(f"{BRONZE_PATH}/product_category_name_translation.csv", header=True, inferSchema=True)

print("Bronze data loaded!")
```

```python
# Clean all datasets

from pyspark.sql.functions import to_timestamp, col, expr

# customers
df_customers_clean = df_customers.dropDuplicates()
df_customers_clean = df_customers_clean.dropna(subset=["customer_id"])

# orders
df_orders_clean = df_orders.dropDuplicates()
df_orders_clean = df_orders_clean.dropna(subset=["order_id"])

# order_items
df_order_items_clean = df_order_items.dropDuplicates()
df_order_items_clean = df_order_items_clean.dropna(subset=["order_id"])

# payments
df_payments_clean = df_payments.dropDuplicates()
df_payments_clean = df_payments_clean.dropna(subset=["order_id"])

# reviews
df_reviews_clean = df_reviews.dropDuplicates()
df_reviews_clean = df_reviews_clean.dropna(subset=["review_id", "order_id"])
df_reviews_clean = df_reviews_clean.withColumn("review_score", expr("try_cast(review_score as double)"))
df_reviews_clean = df_reviews_clean.withColumn("review_creation_date", expr("try_cast(review_creation_date as timestamp)"))
df_reviews_clean = df_reviews_clean.withColumn("review_answer_timestamp", expr("try_cast(review_answer_timestamp as timestamp)"))

# products
df_products_clean = df_products.dropDuplicates()
df_products_clean = df_products_clean.dropna(subset=["product_id"])

# sellers
df_sellers_clean = df_sellers.dropDuplicates()
df_sellers_clean = df_sellers_clean.dropna(subset=["seller_id"])

# geolocation
df_geolocation_clean = df_geolocation.dropDuplicates()

# category
df_category_clean = df_category.dropDuplicates()

print("Data cleaned successfully!")
```

```python
# Compare row counts before and after cleaning

print("ROW COUNTS: Before vs After cleaning ")
print("customers: ", df_customers.count(), "->", df_customers_clean.count())
print("orders: ", df_orders.count(), "->", df_orders_clean.count())
print("order_items: ", df_order_items.count(), "->", df_order_items_clean.count())
print("payments: ", df_payments.count(), "->", df_payments_clean.count())
print("reviews: ", df_reviews.count(), "->", df_reviews_clean.count())
print("products: ", df_products.count(), "->", df_products_clean.count())
print("sellers: ", df_sellers.count(), "->", df_sellers_clean.count())
print("geolocation: ", df_geolocation.count(), "->", df_geolocation_clean.count())
print("category: ", df_category.count(), "->", df_category_clean.count())
```

```python
# Save cleaned data to Silver layer

SILVER_PATH = "/Volumes/olist_project/silver/clean_data"

df_customers_clean.write.mode("overwrite").parquet(f"{SILVER_PATH}/customers")
df_orders_clean.write.mode("overwrite").parquet(f"{SILVER_PATH}/orders")
df_order_items_clean.write.mode("overwrite").parquet(f"{SILVER_PATH}/order_items")
df_payments_clean.write.mode("overwrite").parquet(f"{SILVER_PATH}/payments")
df_reviews_clean.write.mode("overwrite").parquet(f"{SILVER_PATH}/reviews")
df_products_clean.write.mode("overwrite").parquet(f"{SILVER_PATH}/products")
df_sellers_clean.write.mode("overwrite").parquet(f"{SILVER_PATH}/sellers")
df_geolocation_clean.write.mode("overwrite").parquet(f"{SILVER_PATH}/geolocation")
df_category_clean.write.mode("overwrite").parquet(f"{SILVER_PATH}/category")

print("Silver layer saved successfully!")
```
