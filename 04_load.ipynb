```python
# Notebook 04 - Load
# Description: Save Gold tables as Delta Tables in Unity Catalog

GOLD_PATH = "/Volumes/olist_project/gold/star_schema"

dim_customers = spark.read.parquet(f"{GOLD_PATH}/dim_customers")
dim_products = spark.read.parquet(f"{GOLD_PATH}/dim_products")
dim_sellers = spark.read.parquet(f"{GOLD_PATH}/dim_sellers")
dim_dates = spark.read.parquet(f"{GOLD_PATH}/dim_dates")
fact_orders = spark.read.parquet(f"{GOLD_PATH}/fact_orders")

print("Gold data loaded!")
```

```python
# Save as Delta Tables

dim_customers.write.format("delta").mode("overwrite").saveAsTable("olist_project.gold.dim_customers")
dim_products.write.format("delta").mode("overwrite").saveAsTable("olist_project.gold.dim_products")
dim_sellers.write.format("delta").mode("overwrite").saveAsTable("olist_project.gold.dim_sellers")
dim_dates.write.format("delta").mode("overwrite").saveAsTable("olist_project.gold.dim_dates")
fact_orders.write.format("delta").mode("overwrite").saveAsTable("olist_project.gold.fact_orders")

print("Delta Tables saved successfully!")
```
